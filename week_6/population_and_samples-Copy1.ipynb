# Population and samples in statistics

## Libraries and settings

```python
# Libraries
import numpy as np
import random
import statistics
import matplotlib.pyplot as plt

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

##  Prices of mobile phones in the city of ZÃ¼rich (population)

```python
# Setting a seed to restore the random numbers later
np.random.seed(10)

# Generate (normally distributed) data with mean = 560.- and standard deviation 10.-
pop = np.random.normal(loc=560, scale=150, size=150)
print(f'{len(pop):.0f}')
print(f'{np.mean(pop):.4f}')
print(f'{np.std(pop):.4f}')

# Plot histogram based on the data
import matplotlib.pyplot as plt
plt.hist(pop, bins = 50, color='green')
plt.title('Prices of mobile phones in the city of Zuerich')
plt.xlabel('Price')
plt.ylabel('Number of phones')

plt.show()
```

## Single random sample from the population

```python
# Sample size
sample_size = 30

# Setting a seed to restore the random numbers later
random.seed(10)

# Random sample
random_sample = random.sample(list(pop), sample_size)

# Statistics of the sample
print(f'Mean: {statistics.mean(random_sample):.4f}')
print(f'Standard deviation: {statistics.stdev(random_sample):.4f}')

# Plot histogram based on the sample
import matplotlib.pyplot as plt
plt.hist(random_sample, bins = 50, color='deepskyblue')
plt.title(f'Prices of mobile phones in the city of Zuerich (sample with n={sample_size})', fontsize=10)
plt.xlabel('Price')
plt.ylabel('Number of phones')

plt.show()
```

## Repeated random sampling 'with replacement' (bootstrapping)

```python
# Define number of samples and sample size
num_iterations = 50
sample_size = 30

# Create empty lists to save results of iterations
sample_mean = []
sample_std  = []

# Loop for iterations
for i in range(num_iterations):
    samp = random.sample(pop.tolist(), sample_size)
    avg = np.mean(samp)
    std = np.std(samp)
    sample_mean.append(avg)
    sample_std.append(std)

# Print mean and sd of bootsstapped distribution
print(f'{np.mean(sample_mean):.4f}')
print(f'{np.mean(sample_std):.4f}')

# Plot histogram based on the bootstrapped data
plt.hist(sample_mean, bins = 50, color='orange')
plt.title('Prices of mobile phones in the city of Zuerich (bootstrap sample)', fontsize=10)
plt.xlabel('Price')
plt.ylabel('Number of phones')
plt.show()
```
